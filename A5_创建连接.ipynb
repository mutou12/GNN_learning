# 导入工具包

```python
import networkx as nx
import numpy as np
import random
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

# 创建图

```python
G = nx.Graph()
print(G.is_directed())
```

```python
G.graph['Name'] = 'HelloWord'
print(G.graph)
```

```python
H = nx.DiGraph()
print(H.is_directed())
```

# 创建单个节点

```python
G.add_node(0, feature=5, label=0, zihao=2)
```

```python
G.nodes[0]
```

# 创建多个节点

```python
G.add_nodes_from([
    (1, {'feature': 1, 'label': 1, 'zihao': 3}),
    (2, {'feature': 2, 'label': 2, 'zihao': 4})
])
```

# 全图节点信息

```python
G.number_of_nodes()
```

```python
G.nodes
```

```python
G.nodes(data=True)
```

# 创建连接，设置属性特征

```python
G.add_edge(0, 1, weight=0.5, like=3)
```

```python
nx.draw(G)
```

# 创建多个连接

```python
G.add_edges_from([
    (1, 2, {'weight': 0.3, 'like': 5}),
    (2, 0, {'weight': 0.1, 'like': 8})
])
```

```python
G.edges[1, 2]
```

```python
nx.draw(G)
```

# 全图连接信息

```python
G.number_of_edges()
```

```python
G.size()
```

```python
G.edges()
```

```python
G.edges(data=True)
```

```python
for edge in G.edges(data=True):
    print(edge)
```

# 节点的连接数

```python
node_id = 1
```

```python
G.degree[node_id]
```

```python
for neighbor in G.neighbors(node_id):
    print(f'Node {node_id} has neighbor {neighbor}')
```
