# 导入工具包

```python
import networkx as nx
import numpy as np
import random
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

# 创建4x4网格图

```python
G = nx.grid_2d_graph(4, 4)
```

# 原生可视化

```python
pos = nx.spring_layout(G, seed=123)
nx.draw(G, pos)
```

# 不显示节点

```python
nx.draw(G, pos, node_size=0, with_labels=False)
```

# 设置颜色

```python
len(G.edges())
```

```python
nx.draw(
    G,
    pos,
    node_color='#A0CBE2',
    edgecolors='red',
    edge_color='yellow',
    node_size=800,
    with_labels=False,
    width=3
)
```

# 有向图

```python
nx.draw(
    G.to_directed(),
    pos,
    node_color='tab:pink',
    node_size=400,
    with_labels=False,
    edgecolors='tab:blue',
    arrowsize=10,
    width=2
)
```

# 设置每个节点的坐标
## 无向图

```python
G = nx.Graph()
G.add_edges_from([
    (1, 2),
    (1, 3),
    (1, 5),
    (2, 3),
    (3, 4),
    (4, 5)
])
```

```python
nx.draw(G, with_labels=True)
```

```python
pos = {
    1:(0, 0),
    2:(-1, 0.3),
    3:(2, 0.17),
    4:(4, 0.255),
    5:(5, 0.03)
}
options = {
    'font_size': 36,
    'node_size': 3000,
    'node_color': 'white',
    'edgecolors': 'black',
    'linewidths': 5,
    'width': 5
}
nx.draw_networkx(G, pos, **options)
ax = plt.gca()
ax.margins(0.20)
plt.axis('off')
plt.show()
```

# 有向图

```python
G = nx.DiGraph([(0, 3), (1, 3), (2, 4), (3, 5), (3, 6), (4, 6), (5, 6)])
```

```python
nx.draw(G, with_labels=True)
```

```python
left_nodes = [0, 1, 2]
middle_nodes = [3, 4]
right_nodes = [5, 6]
```

```python
pos = {n: (0, i) for i, n in enumerate(left_nodes)}
pos.update({n: (1, i+0.5) for i, n in enumerate(middle_nodes)})
pos.update({n: (2, i+0.5) for i, n in enumerate(right_nodes)})
```

```python
pos
```

```python
nx.draw_networkx(G, pos, **options)

ax = plt.gca()
ax.margins(0.20)
plt.axis('off')
plt.show()
```

```python
nx.draw_networkx_nodes(G, pos, node_size=3000, nodelist=[0, 1, 2, 3], node_color='tab:blue')
nx.draw_networkx_nodes(G, pos, node_size=2000, nodelist=[4, 5, 6], node_color='tab:orange')
nx.draw_networkx_edges(G, pos, alpha=0.5, width=6)
plt.axis('off')
plt.show()
```
