# 导入工具包

```python
import networkx as nx
import numpy as np
import random
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

# 导入三元组连接表

```python
df = pd.read_csv('data/data/三国演义/triples.csv')
```

```python
df
```

# 通过连接表Edge List创建图

```python
G = nx.DiGraph()
```

```python
edges = [edge for edge in zip(df['head'], df['tail'])]
G.add_edges_from(edges)
```

```python
G.edges('关羽')
```

# 可视化

```python
pos = nx.spring_layout(G, seed=123)

plt.figure(figsize=(15, 15))
nx.draw(G, pos=pos, with_labels=True)
```

# 查看全图参数

```python
print(G)
```

```python
G.nodes
```

# 保存并载入邻接表Adjacency List

```python
for line in nx.generate_adjlist(G):
    print(line)
```

```python
nx.write_edgelist(G, path='data/data/三国演义/grid.edgelist', delimiter=":")
```

```python
H = nx.read_edgelist(path='data/data/三国演义/grid.edgelist', delimiter=":")
```

```python
pos = nx.spring_layout(H, iterations=3, seed=5)

plt.figure(figsize=(15, 15))
nx.draw(G, pos=pos, with_labels=True)
plt.show()
```
