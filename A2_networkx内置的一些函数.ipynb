# 经典图结构

## 全连接无向图

```python
G = nx.complete_graph(7)
```

```python
nx.draw(G)
```

```python
G.size()
```

## 全连接有向图

```python
G = nx.complete_graph(7, nx.DiGraph())
nx.draw(G)
```

## 环状图

```python
G = nx.cycle_graph(5)
nx.draw(G)
```

## 梯状图

```python
G = nx.ladder_graph(5)
nx.draw(G)
```

## 线性串珠图

```python
G = nx.path_graph(15)
nx.draw(G)
```

## 星状图

```python
G = nx.star_graph(7)
nx.draw(G)
```

## 轮廓图

```python
G = nx.wheel_graph(8)
nx.draw(G)
```

## 二项树

```python
G = nx.binomial_tree(5)
nx.draw(G)
```

# 栅格图

# 社交网络

```python
G = nx.karate_club_graph()
nx.draw(G, with_labels=True)
```

```python
G.nodes[5]["club"]
```

```python
G.nodes[9]["club"]
```

```python
G = nx.les_miserables_graph()
plt.figure(figsize=(12, 10))
pos = nx.spring_layout(G, seed=7)
nx.draw(G, pos, with_labels=True)
```

```python
G = nx.florentine_families_graph()
nx.draw(G, with_labels=True)
```
