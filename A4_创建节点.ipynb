# 导入工具包

```python
import networkx as nx
import numpy as np
import random
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

# 创建无节点、无连接的空图

```python
G = nx.Graph()
```

```python
G
```

```python
nx.draw(G)
```

# 添加单个节点

```python
G.add_node('刘备')
```

```python
G.nodes
```

# 添加多个节点

```python
G.add_nodes_from([1,2,3,4])
```

```python
G.nodes
```

# 添加带属性特征的节点

```python
G.add_nodes_from([
    ('关羽', {'武器': '青龙偃月刀', '武力值': 90, '智力值': 80}),
    ('张飞', {'武器': '丈八蛇矛', '武力值': 85, '智力值': 75}),
    ('吕布', {'武器': '方天画戟', '武力值': 100, '智力值': 70})
])
```

```python
G.nodes
```

```python
nx.draw(G, with_labels=True)
```

# 创建另一个首尾相连成串的Path Graph

```python
H = nx.path_graph(10)
```

```python
nx.draw(H)
```

```python
H.nodes
```

# 将H的节点添加到G中

```python
G.add_nodes_from(H)
```

```python
G.nodes
```

```python
nx.draw(G)
```
